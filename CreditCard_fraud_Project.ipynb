# Dropping Duplicate Records

```python
df.drop_duplicates(inplace=True)
```

```python
# Calculating the number of fraud cases in this dataset
fraud = df[df['Class'] == 1]
valid = df[df['Class'] == 0]
fraction = len(fraud)/float(len(valid))

print(fraction)
print("Total No. of Fraud Cases: {}".format(len(df[df['Class'] == 1])))
print("Total No. of Valid Cases: {}".format(len(df[df['Class'] == 0])))
```
